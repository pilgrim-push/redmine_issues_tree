```
# System repositories ship PostGIS that's too old. Install hot new one from PGDG:
!curl https://www.postgresql.org/media/keys/ACCC4CF8.asc | apt-key add -
!sh -c 'echo "deb http://apt.postgresql.org/pub/repos/apt/ bionic-pgdg main" > /etc/apt/sources.list.d/pgdg.list'
!apt update

!apt install postgis xxd

# Tweak default install to be usable in Colab environment where machine is throwaway and keyboardless
!echo "local all  all trust" > /etc/postgresql/12/main/pg_hba.conf

# disable pager, jupyter is our pager
!echo "\pset pager off" >> ~/.psqlrc 

# set borders to cool ├── ones instead of 1960 +--- ones
!echo "\pset linestyle unicode" >> ~/.psqlrc # 
!echo "\pset border 2" >> ~/.psqlrc

!echo "SET postgis.gdal_enabled_drivers = 'ENABLE_ALL';" >> ~/.psqlrc
%env PGDATABASE=postgres
%env PGUSER=postgres

# Start Postgres and install PostGIS into system database.
!service postgresql start
!psql -c "create extension postgis;"
!psql -c "create extension postgis_raster;"
from IPython.display import Image
```
